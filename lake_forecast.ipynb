```r
library(forecast)
time_series <- ts(df$Temperature_Le_Croci,start = 2004,frequency = 365 )
```

```r
plot(time_series)
acf(time_series, lag.max = 36)
pacf(time_series,lag.max = 36)
```

```r
components = decompose(time_series)
plot(components)
```

```r
install.packages('fUnitRoots', repos='http://cran.us.r-project.org')
```

```r
library("fUnitRoots")
urkpssTest(time_series, type = c("tau"), lags = c("short"),use.lag = NULL, doplot = TRUE)
tsstationary = diff(time_series, differences=1)
plot(tsstationary)
```

```r
acf(time_series,lag.max=34)
```

```r
timeseriesseasonallyadjusted <- time_series- components$seasonal
tsstationary <- diff(timeseriesseasonallyadjusted, differences=1)
plot(tsstationary)
```

```r
acf(tsstationary, lag.max=34)
pacf(tsstationary, lag.max=34)
```

```r
fitARIMA <- arima(time_series, order=c(1,1,1),seasonal = list(order = c(1,0,0), period = 12),method="ML")
library(lmtest)
coeftest(fitARIMA)
```
